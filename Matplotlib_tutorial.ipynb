```python
import numpy as np
a = np.arange(10)
```

```python
plt.plot(a,a,'ro',a,a**2,'bs',a,a**3,'g^')
plt.show()
```

```python
t1 = np.arange(0,17,0.5)
t2 = np.arange(0,17,0.2)
plt.figure(1)
plt.subplot(211)
plt.plot(t1,t1**3,'bo',t2,t2**3,'k')
plt.subplot(212)
plt.plot(t1,t1**2,'bo',t2,t2**2,'k')
plt.show()
```
